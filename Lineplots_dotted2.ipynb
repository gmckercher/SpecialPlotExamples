```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
def incoming(lat):
    
    print 'latitude', lat*(180/np.pi)
    
    # Given Values
    T_sunrise = 280 #K
    S = 1365 #W/m^2
    tau = 0.3
    Csoil = 2*math.pow(10,6) #J/m^2K
    e_sw = 1
    e_lw = 0.9
    sigma = 5.67*math.pow(10,-8) #W/m^2K^4
    dec = 23.5*(np.pi/180) #rad
    
    # 30N Hour Angle (Radians)
    if abs(lat) <= (65*np.pi/180):
        H_sunset = np.arccos(-np.tan(lat)*np.tan(dec)) 
        H_sunrise = -H_sunset
        print 'H_sunrise', H_sunrise
        print 'H_sunset', H_sunset

        # Length of Day (hours)
        LoD = (2*H_sunset)/(np.pi/12)
        print 'Length of Day', LoD
        
        # Change in H/hour
        H_diff = (H_sunset - H_sunrise)/LoD
        H = [0]*(int(LoD)+1)
        H[0] = H_sunrise
        for x in range(1,int(LoD)):
            H[x] = H[x-1] + H_diff
        H[(int(LoD))] = H_sunset
        print 'Hour Angle'
        print np.around(H,2)
        
        # Zenith Angle
        theta = []
        for x in H:
            theta.append(np.arccos((np.sin(lat)*np.sin(dec))+(np.cos(lat)*np.cos(dec)*np.cos(x))))
        print 'Zenith Angle'
        print np.around(theta,2)
        
    else:
        LoD = 24
        F = np.genfromtxt('zenithfinland.txt', usecols=(0))
        theta = F*(np.pi/180) #rad
        print 'Zenith Angle'
        print np.around(theta,2)
        
        L = np.genfromtxt('zenithfinland.txt', usecols=(1))
        H = L*(np.pi/180) #rad
        print 'Hour Angle'
        print np.around(H,2)


    # incoming solar irradiance
    income = np.abs(S*np.exp(-tau/np.cos(theta)))
    print 'income'
    print np.around(income,2)
    
    return income, LoD
```

```python
lat70N = 70*(np.pi/180) #rad
lat65N = 65*(np.pi/180) #rad
lat60N = 60*(np.pi/180) #rad
lat30N = 30*(np.pi/180) #rad
lat0 = 1*(np.pi/180) #rad
lat30S = -30*(np.pi/180) #rad
lat60S = -60*(np.pi/180) #rad
lat89S = -60*(np.pi/180) #rad
```

```python
income30N,LoD30 = incoming(lat30N)
income60N,LoD60 = incoming(lat60N)
income65N,LoD65 = incoming(lat65N)
#income70N,LoD70 = incoming(lat70N)
income0,LoD0 = incoming(lat0)

plt.figure(figsize=[15,6])

plt.scatter(np.arange(0,LoD0),income0,s=30)
plt.plot(np.arange(0,LoD0),income0,label = 'Equator', linewidth=6)

plt.scatter(np.arange(0,LoD30),income30N,s=30)
plt.plot(np.arange(0,LoD30),income30N, label = '30N',linestyle='-',linewidth=3)

plt.scatter(np.arange(0,LoD60),income60N,s=30)
plt.plot(np.arange(0,LoD60),income60N, label = '60N',linestyle='-.',linewidth=4)

plt.scatter(np.arange(0,LoD65),income65N,s=30)
plt.plot(np.arange(0,LoD65),income65N,linestyle='--',label = '65N', linewidth=4)




plt.xlabel('Hours',fontsize=16)
plt.ylabel('Incoming Irradiance (W/m^2)',fontsize=16)
plt.ylim(0,max(income)+100)
plt.xlim(0,25)
plt.legend()
plt.show()
```
