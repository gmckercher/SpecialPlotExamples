```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('project_data_reduced.csv')
```

```python
# credit to http://chrisalbon.com/python/pandas_resample_by_time.html as a resource
df.index = pd.date_range('5/12/2016', periods=38400, freq='500mS')
```

```python
mean = df.resample('600S', how='mean')
```

```python
std = df.resample('600S', how='std')
```

```python
plt.figure(figsize=(12,6))
plt.title('Mean Turbulence Intensity Comparison')
plt.ylabel('Turbulence Intensity')
plt.plot(std['FTWS']/mean['FTWS'],label='FTWS')
plt.plot(std['PVWS']/mean['PVWS'],label='PVWS')
plt.plot(std['SONWS']/mean['SONWS'],label='SONWS')
plt.legend()
```

```python
print 'Mean TI FTWS =', np.mean(std['FTWS']/mean['FTWS'])
print 'Mean TI PVWS =', np.mean(std['PVWS']/mean['PVWS'])
print 'Mean TI SONWS =', np.mean(std['SONWS']/mean['SONWS'])
```
