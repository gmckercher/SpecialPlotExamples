## Functions

```python
def calctheta(dx,dy):
    'Calculates theta - northrelative azimuth'
    theta = math.atan2(dx,dy) #radians
    if theta < 0.0:
        theta = theta + (2*np.pi)
    return theta
    
    
    # accounting for the change in angle with quadrant
    '''
    if x>=0 & y>0:
        #theta = np.degrees(np.arctan(dx/dy)) #degrees
        theta = np.arctan(dx/dy) #radians
        return theta
    elif x>0 & y<=0:
        #theta = np.degrees(np.arctan(dy/dx)) + 90 #degrees
        theta = np.arctan(dy/dx) + (np.pi/2) #radians
        return theta
    elif x<=0 & y<0:
        #theta = np.degrees(np.arctan(dx/dy)) + 180 #degrees
        theta = np.arctan(dx/dy) + np.pi #radians
        return theta
    elif x<0 & y>=0:
        #theta = np.degrees(np.arctan(dy/dx)) + 270 #degrees
        theta = np.arctan(dy/dx) + ((3*np.pi)/2) #radians
        return theta
    elif x==0 & y==0:
        theta = 0
        return theta
    '''
    
def calcphi(x,xo,y,yo,z,zo):
    'Calculates phi - elevation angle'
    dy = np.abs(float(yo-y))
    dx = np.abs(float(xo-x))
    dz = np.abs(float(zo-z))
    H = np.sqrt((dx**2)+(dy**2))
    if H==0:
        phi = 0
    else:
        phi = np.arctan(dz/H)
    return phi

def calcu(Vr1,Vr2,phi1,phi2,theta1,theta2):
    'Calculates wind in u direction'
    denom = (np.cos(phi1)*np.cos(phi2)*np.sin(theta1-theta2))
    u = (((Vr1*np.cos(phi2)*np.cos(theta2)) - (Vr2*np.cos(phi1)*np.cos(theta1))) / (denom))
    return u
    
def calcv(Vr1,Vr2,phi1,phi2,theta1,theta2):
    'Calculates wind in u direction'
    denom = (np.cos(phi1)*np.cos(phi2)*np.sin(theta1-theta2))
    v = (((Vr2*np.cos(phi1)*np.sin(theta1)) - (Vr1*np.cos(phi2)*np.sin(theta2))) / (denom))
    return v

def calcw(u,v,Dx,Dy,Dz):
    'Calculates wind in w direction with u,v wind and grid spacing'
    for x in range(258):
        for y in range(258): 
            for z in range(110):
                
                # Finite Differencing Method for dw/dz
                
                if x==0 & y==0:
                    dudx = ((u[z,y,x+1])-(u[z,y,x])) / Dx
                    dvdy = ((v[z,y+1,x])-(v[z,y,x])) / Dy
                elif x==258 & y==258:
                    dudx = ((u[z,y,x])-(u[z,y,x-1])) / Dx
                    dvdy = ((v[z,y,x])-(v[z,y-1,x])) / Dy
                else:
                    dudx = ((u[z,y,x+1])-(u[z,y,x-1])) / (2*Dx)
                    dvdy = ((v[z,y+1,x])-(v[z,y-1,x])) / (2*Dx)
                
                # Estimation for dw
                
                dwdz = -1*(dudx+dvdy)
                if z==0:
                    w[z,y,x] = dwdz * Dz
                else:
                    w[z,y,x] = w[z,y,x-1] + dwdz
                
                return w
```

```python
rad1_elev = np.zeros([110,258,258])
rad1_azi = np.zeros([110,258,258])
rad2_elev = np.zeros([110,258,258])
rad2_azi = np.zeros([110,258,258])
u1 = np.zeros([110,258,258])
v1 = np.zeros([110,258,258])
w = np.zeros([110,258,258])

xo1,yo1,zo1 = np.array([0,124,0])
xo2,yo2,zo2 = np.array([257,124,0])
```

### Looping

```python
for x in range(258):
    for y in range(258): 
        for z in range(110):
            
            dx1 = -(xo1-x)
            dx2 = -(xo2-x)
            dy1 = -(yo1-y)
            dy2 = -(yo2-y)
            dz1 = -(zo1-z)
            dz2 = -(zo2-z)
            rad1_azi[z,y,x] = calctheta(dx1,dy1)
            rad2_azi[z,y,x] = calctheta(dx2,dy2)
            rad1_elev[z,y,x] = calcphi(x,xo1,y,yo1,z,zo1)
            rad2_elev[z,y,x] = calcphi(x,xo2,y,yo2,z,zo2)

print np.min(rad1_azi),np.max(rad1_azi)
print np.min(rad2_azi),np.max(rad2_azi)
print np.min(rad1_elev),np.max(rad1_elev)
print np.min(rad2_elev),np.max(rad2_elev)
```

```python
u1 = calcu(prad1,prad2,rad1_elev,rad2_elev,rad1_azi,rad2_azi)
v1 = calcv(prad1,prad2,rad1_elev,rad2_elev,rad1_azi,rad2_azi)
print u1[0,:,:]
```
