```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ke = 2.3e-4 # mass abs cross section
ke2 = ke*100
rho = 1.2
Tv = 250
R = 287
g = 9.8
```

```python
H = (R*Tv)/g
z = np.arange(0,100*(10**3),1**3)
print len(z)
```

```python
A = [0]*100000
t = [0]*100000
W = [0]*100000
Wn = [0]*100000
A2 = [0]*100000
t2 = [0]*100000
W2 = [0]*100000
Wn2 = [0]*100000
for i in z:
    A[i] = (ke*rho*H*(np.exp(-z[i]/H)))
    t[i] = (np.exp(-A[i]))
    W[i] = (ke*rho*np.exp(-(A[i])-(z[i]/H)))
#print max(W), min(W) 
    mi = 3.22953729033e-10
    ma = 5.02469480291e-05
    Wn[i] = (W[i]-mi) / (ma-mi)


    # ke 100 times larger
    A2[i] = (ke2*rho*H*(np.exp(-z[i]/H)))
    t2[i] = (np.exp(-A2[i]))
    W2[i] = (ke2*rho*np.exp(-(A2[i])-(z[i]/H)))
#print max(W2), min(W2)
    mi2 = 4.81285417531e-90
    ma2 = 5.02469480178e-05 
    Wn2[i] = (W2[i]-mi2) / (ma2-mi2)
```

```python
plt.figure(figsize=[10,10])
plt.plot(t,z,c='b',linewidth = 4,label = 't1')
plt.plot(Wn,z,c='b',linestyle='--',linewidth = 4,label = 'W1')
plt.plot(t2,z,c='r',linewidth = 2,label = 't2')
plt.plot(Wn2,z,c='r',linestyle='-.',linewidth = 2,label = 'W2')
plt.xlabel('VALUE 0 TO 1',fontsize=16)
plt.ylabel('Z (METERS)',fontsize=16)
plt.ylim(0,100000)
plt.xlim(0,1)
plt.legend()
plt.show()
```
