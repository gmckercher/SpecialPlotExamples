# part 1b contour the solution 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = z = np.arange(0,41,1) # array from -10 to 10
X,Z = np.meshgrid(x,z) # making a grid
print x
```

```python
xo = 20
zo = 20
q = 0.5
dist = np.sqrt((xo-X)**2+(zo-Z)**2)
phi = q/((4*np.pi)*np.abs(dist))
```

```python
print phi[18,20]
print phi[19,20]
print phi[20,20]
phi[20,20] = (phi[19,20]-phi[18,20])+phi[19,20]
```

```python
print X.shape
print Z.shape
print phi.shape
```

```python
fig = plt.figure(figsize=[8,6])
plt.title('Solution for $\Phi$ on the (x,z) plane', fontsize=16)
plot = plt.contourf(X,Z,phi);
plt.xlabel('X')
plt.ylabel('Z')
plt.colorbar(label='$\Phi$')
```

```python
fig = plt.figure(figsize=[6,6])
plt.quiver(X,Z,0,-phi)
plt.title('Vertical Velocity Vectors', fontsize=16)
plt.xlabel('X')
plt.ylabel('Z')
#plt.colorbar()
#plt.contour(X,Z,phi);
```
