```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
f = np.genfromtxt('springcleaned.csv',delimiter=',',skiprows=3)
O3_ppm = f[:,0]
NO2_ppm = f[:,1]
SO2_ppm = f[:,2]
CO_ppm = f[:,3]
VOC_ppm = f[:,4]
NMHC_ppm = f[:,5]
```

```python
O3 = pd.DataFrame(O3_ppm)
NO2 = pd.DataFrame(NO2_ppm)
SO2 = pd.DataFrame(SO2_ppm)
CO = pd.DataFrame(CO_ppm)
VOC = pd.DataFrame(VOC_ppm)
NMHC = pd.DataFrame(NMHC_ppm)
```

```python
def box_array(array):
    stat = array.describe()
    box = np.array(stat)
    num = float(box[3]),float(box[4]),float(box[5]),float(box[6]),float(box[7])
    return num
```

```python
plt.figure(figsize=[16,10])

ax1 = plt.subplot(6,1,1)
ax1.set_title('Fixed Site Statistics (April 5 to May 8, 2015)',fontsize=14)
ax1.boxplot([box_array(O3)],vert=False,widths=0.6)
#ax1.boxplot([0.041754],vert=False,showmeans=True)
ax1.scatter(0.075,1,marker='d',s=100,c='yellow')
ax1.set_ylabel('O3',fontsize=14)
ax1.set_yticks([])

ax2 = plt.subplot(6,1,2)
ax2.boxplot([box_array(NO2)],vert=False,labels=['NO2'],widths=0.6)
#ax2.boxplot([0.055335],vert=False,showmeans=True)
ax2.scatter(0.1,1,marker='d',s=100,c='yellow')
ax2.set_ylabel('NO2',fontsize=14)
ax2.set_yticks([])

ax3 = plt.subplot(6,1,3)
ax3.boxplot([box_array(SO2)],vert=False,labels=['SO2'],widths=0.6)
#ax3.boxplot([0.064342],vert=False,showmeans=True)
ax3.scatter(0.075,1,marker='d',s=100,c='yellow')
ax3.scatter(0.5,1,marker='d',s=100,c='yellow')
ax3.set_ylabel('SO2',fontsize=14)
ax3.set_xlim(0,0.8)
ax3.set_yticks([])

ax4 = plt.subplot(6,1,4)
ax4.boxplot([box_array(CO)],vert=False,labels=['CO'],widths=0.6)
#ax4.boxplot([0.788924],vert=False,showmeans=True)
ax4.scatter(9,1,marker='d',s=100,c='yellow')
ax4.set_ylabel('CO',fontsize=14)
ax4.set_xlim(-0.2,12)
ax4.set_yticks([])

ax5 = plt.subplot(6,1,5)
ax5.boxplot([box_array(VOC)],vert=False,labels=['VOC'],widths=0.6)
#ax5.boxplot([0.312743],vert=False,showmeans=True)
ax5.set_ylabel('VOC',fontsize=14)
ax5.set_yticks([])

ax6 = plt.subplot(6,1,6)
ax6.boxplot([box_array(NMHC)],vert=False,labels=['NMHC'],widths=0.6)
#ax6.boxplot([0.237337],vert=False,showmeans=True)
ax6.set_xlabel('Concentration (ppm)',fontsize=14)
ax6.set_ylabel('NMHC',fontsize=14)
ax6.set_yticks([])
```

```python
def myround(a, decimals=1):
     return np.around(a-10**(-(decimals+5)), decimals=decimals)
```
