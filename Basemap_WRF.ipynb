### Let's plot a skew-T

```python
from mpl_toolkits.axisartist import Subplot
from matplotlib.ticker import FuncFormatter, Formatter 
from mpl_toolkits.axisartist.grid_helper_curvelinear import GridHelperCurveLinear 
```

```python
C_to_K = 273.15
skew_slope = 40

def x_from_Tp(Tp, p):
    """ Transforms the x-coordinate based on temperature and the log of pressure """
    x = Tp - skew_slope * np.log(p)
    return x

def y_from_p(p):
    """ Transforms the y coordinate based on the negative of pressure """
    y = -np.log(p)
    return y

def p_from_y(y):
    """ Transforms from the y coordinate back into pressure """
    p = np.exp(-y)
    return p

def T_from_xp(x, p):
    """ Transforms from the x coordinate and a given pressure back to temperature """
    T = x + skew_slope * np.log(p)
    return T

def to_thermo(x, y):
    """ Transform (x,y) coordinates to T in degrees Celsius and p in mb. """
    p = p_from_y(y)
    T_C = T_from_xp(x, p) - C_to_K
    return T_C, p

def from_thermo(T_C, p):
    """ Transform T_C (in degrees Celsius) and p (in mb) to (x,y). """
    y = y_from_p(p)
    x = x_from_Tp(T_C + C_to_K, p)
    return x, y
```

```python
desired_lat = -102.2
desired_lon = 33.3
time=24

xs = lons[lat_min:lat_max,selects_lon]
ys = lats[lat_min:lat_max,selects_lon]

length = np.shape(xs)[1]
nth = np.argmin((xs-desired_lat)**2+(ys-desired_lon)**2)
grid_number = nth/length, nth%length
```

```python
xl = grid_number[0]
yl = grid_number[1]
snd_T = t[time,:,xl,yl] - C_to_K
good_T = (snd_T > -100.0) & (snd_T < 60.0) # all temperature values, deg.C, should be in this range.
snd_Td = td[time,:,xl,yl]
good_Td = (snd_Td > -100.0) & (snd_Td < 60.0) # all dewpoint values, deg. C, should be in this range.
snd_p = p[time,:,xl,yl]
good_p = (snd_p > 0.0) & (snd_p < 1500.0) # all pressure values, mb, should be in this range.

x_snd_Td = x_from_Tp(snd_Td[good_Td & good_p & good_T]+C_to_K, snd_p[good_Td & good_p & good_T])
x_snd_T  = x_from_Tp(snd_T[good_Td & good_p & good_T] +C_to_K, snd_p[good_Td & good_p & good_T])
y_snd_p  = y_from_p(snd_p[good_Td & good_p & good_T])
```

```python
# values along the bottom and left edges
p_bottom = 950.0
p_top    = 150
T_min    = -15 + C_to_K
T_max    =  40 + C_to_K

# Converting bottom and left edges from T (K) and p (mb) to (x,y)
x_min = x_from_Tp(T_min, p_bottom)
x_max = x_from_Tp(T_max, p_bottom)
y_min = y_from_p(p_bottom)
y_max = y_from_p(p_top)

p_levels   = np.arange(1000, 150-50, -50)
T_C_levels = np.arange(-80, 40+10, 10)
T_levels   = T_C_levels + C_to_K
theta_levels    = np.arange(-40, 100+10, 10) + C_to_K
theta_ep_levels = theta_levels.copy()
mixing_ratios   = np.asarray((0.4, 1, 2, 3, 5, 8, 12, 16, 20)) / 1000

p_all       = np.arange(p_bottom, p_top-1, -1)
y_p_levels  = y_from_p(p_levels)
y_all_p     = y_from_p(p_all)
x_T_levels  = [x_from_Tp(Ti, p_all) for Ti in T_levels]
x_thetas    = [x_from_Tp(Bolton.theta_dry(theta_i, p_all), p_all) for theta_i in theta_levels]
x_mixing_ratios = [x_from_Tp(Bolton.mixing_ratio_line(p_all, mixing_ratios_i)+C_to_K, p_all) for mixing_ratios_i in mixing_ratios]

mesh_T, mesh_p = np.meshgrid(np.arange(-60.0, T_levels.max()-C_to_K+0.1, 0.1), p_all)
theta_ep_mesh = Bolton.theta_ep_field(mesh_T, mesh_p)

skew_grid_helper = GridHelperCurveLinear((from_thermo, to_thermo))

fig = plt.figure(figsize=(4.5,6))
ax = Subplot(fig, 1, 1, 1, grid_helper = skew_grid_helper)
def format_coord(x,y):
    T,p=to_thermo(x,y)
    return "{0:5.1f} C, {1:5.1f} mb".format(float(T), float(p))

ax.format_coord = format_coord
fig.add_subplot(ax)

for yi in y_p_levels:
    ax.plot((x_min, x_max), (yi, yi), linewidth=0.5, color=(1.0, 0.8, 0.8))

for x_T in x_T_levels:
    ax.plot(x_T, y_all_p, linewidth=0.5, color=(1.0, 0.5, 0.5))

for x_theta in x_thetas:
    ax.plot(x_theta, y_all_p, linewidth=0.1, color=(1.0, 0.7, 0.7))

for x_mixing_ratio in x_mixing_ratios:
    good = p_all >= 600
    ax.plot(x_mixing_ratio[good], y_all_p[good], linewidth=0.3, color=(0.8, 0.8, 0.6))

n_moist = len(theta_ep_levels)
moist_colors = ((0.6, 0.9, 0.7),)*n_moist
ax.contour(x_from_Tp(mesh_T+C_to_K, mesh_p), y_from_p(mesh_p), theta_ep_mesh, theta_ep_levels[:-1], linewidths=0.1, colors = moist_colors)
ax.plot(x_snd_Td, y_snd_p, linewidth=1.5, color='g')
ax.plot(x_snd_T,  y_snd_p, linewidth=1.5, color='r')
ax.axis((x_min, x_max, y_min, y_max))

plt.show()
# plt.savefig('/home/vanna/Desktop/wrf_plots/24hr_%s_%s.png' %(desired_lat, desired_lon), dpi=250)
# plt.clf()
```
